```python
import pandas as pd
import numpy as np
import textwrap
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
df = pd.read_csv(bbc)
```

```python
df.head()
```

```python
doc = df[df['labels']=='business']['text'].sample(random_state = 42)
```

```python
doc
```

```python
def wrap(x):
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
print(wrap(doc.iloc[0]))
# Note that for these articles, the title is included with the article text.
# it's the first line in the text. This is actually quite useful because the title can be thought of as a summary for the text.
# Therefore, one way to check whether or not our summary is good is to see if the title gives us thesame information as our own summary.
```

```python
# we're trying to remove the title from the article before we do any processing. we only want to split once since we only want to remove the title.
# The first of which is the title in the second of which is the article text.
sents = nltk.sent_tokenize(doc.iloc[0].split('\n', 1)[1])
```

```python
# The next step is to create our TF IDF vector riser object. we're using English stop words and also L1 normalization.This ensures that we won't be biased towards longer sentences,
# which would have a higher score simply due to the fact that they contain more words.
featurizer = TfidfVectorizer(stop_words=stopwords.words('english'), norm = 'l1')
```

```python
X = featurizer.fit_transform(sents)
```

```python
def get_sentence_score(tfidf_row):
  # return the average of the non-zero values
  # of the tf-idf vector representation of a sentence
  x = tfidf_row[tfidf_row != 0]
  return x.mean()
```

```python
scores = np.zeros(len(sents))
for i in range(len(sents)):
  score = get_sentence_score(X[i,:])
  scores[i] = score
```

```python
scores
```

```python
sort_idx = np.argsort(-scores)
```

```python
print("Generated summary:")
for i in sort_idx[:5]:
  print(wrap("%.2f: %s" % (scores[i], sents[i])))
```

```python
def summarize(text):
  # extract sentences
  sents = nltk.sent_tokenize(text)

  # perform tf-idf
  X = featurizer.fit_transform(sents)

  # compute scores for each sentence
  scores = np.zeros(len(sents))
  for i in range(len(sents)):
    score = get_sentence_score(X[i,:])
    scores[i] = score
  
  # sort the scores
  sort_idx = np.argsort(-scores)

  # print summary
  for i in sort_idx[:5]:
    print(wrap("%.2f: %s" % (scores[i], sents[i])))
```

```python
doc = df[df.labels == 'entertainment']['text'].sample(random_state=123)
summarize(doc.iloc[0].split("\n", 1)[1])
```
