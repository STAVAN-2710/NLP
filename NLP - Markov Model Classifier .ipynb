```python
input_files = [
  'edgar_allan_poe.txt',
  'robert_frost.txt',
]
```

```python
# collect data into lists
# collect our data into lists for this experiment.We'll be treating each line as an input sample.
input_texts = []
labels = []

#when you loop through a file line by line in Python, It includes the new line character at the end 
#rstrip call will remove the new line since it's not needed.
# The final step is to append this text to our list of input texts and to append the label to our list of labels
for label, f in enumerate(input_files):
    print(f"{f} corresponds to label {label}")
    
    for line in open(f):
        line = line.rstrip().lower() 
        if line:                #to check if an blank line is present or not
            # remove punctuation
            line = line.translate(str.maketrans('', '', string.punctuation))

            input_texts.append(line)
            labels.append(label)
```

```python
train_text, test_text, Ytrain, Ytest = train_test_split(input_texts, labels)
```

```python
# The next step will be to convert our text into integers, as you recall these will be indexes into the markov matrices 
# we are about to define.
# We'll begin by setting idx to one which will act as our current index as we loop through the text.
# We'll also initialize a word to index dictionary with one entry, which maps the unknown token to zero.
# Basically, this will never be used for the train set, but it may be used for the test, If the test set contains words 
# that do not appear in the train set.
```

```python
idx = 1
word2idx = {'<unk>' : 0}
```

```python
# The next step is to loop through our train text in order to populate where idx inside the loop will begin by splitting 
# the text into tokens.
for text in train_text:
    tokens = text.split()
    for token in tokens:
        if token not in word2idx:
            word2idx[token] = idx
            idx += 1
```

```python
# The next step is to convert this data into integer format, this is what we need to index our A our pie.
# first step will be to create two empty lists, which we will use to store the results. loop through the train tracks 
# inside the loop will begin by again splitting the text to do tokenisation. use a list comprehension to map each token
# to its corresponding text.The result of this will be a list of integers representing one line of a poem.

# It's possible that not every word in the test set appears in the trains there, so we can't naively try to index the 
# word to dictionary. we'll use the get function and will ensure that we always return a default value of zero this 
# corresponds to the unknown token.
train_text_int = []
test_text_int = []

for text in train_text:
    tokens = text.split()
    line_as_int = [word2idx[token] for token in tokens]
    train_text_int.append(line_as_int)
    
for text in test_text:
    tokens = text.split()
    line_as_int = [word2idx.get(token, 0) for token in tokens]
    test_text_int.append(line_as_int)
    
```

```python
# we're ready to build our A and pie matrices which represent the markov model.Since we have two classes, 
# we'll have two A's and two pies. The first step is to assign the length of word to idx to a variable called V.
# initialize A0, Pi zero A1 and PI one. initialized each of these arrays to all ones. 
# The reason for this is we're going to be using add one smoothing. So these ones are the initial fake counts 
# for each initial word and each transition. The next step will be to add the rest of the counts from the actual text.

V = len(word2idx)

A0 = np.ones((V,V))
pi0 = np.ones(V)

A1 = np.ones((V,V))
pi1 = np.ones(V)
```

```python
# build a function to create calculate the counts of each transition
# going to take in three arguments the text for the specific class and the pointer to A and pie.
# because A and pi objects, any modifications we made to them inside this function will be remembered outside the function also
# the data we pass in will be of only single class at a time

# compute counts for A and pi

def compute_counts(text_as_int, A, pi):
    for tokens in text_as_int:
        last_idx = None
        for idx in tokens:
            if last_idx == None:
                # it is the first word in the sentence
                pi[idx] += 1
                
            else:
                # the last word exists, so count a transition
                A[last_idx, idx] += 1
            
            # update last idx
            last_idx = idx
            
compute_counts([t for t, y in zip(train_text_int, Ytrain) if y == 0], A0, pi0)
compute_counts([t for t, y in zip(train_text_int, Ytrain) if y == 1], A1, pi1)
```

```python
# at present the A and pi only have counts but we need probability
# normalize A and pi so they are valid probability matrices
# convince yourself that this is equivalent to the formulas shown before
# each row of A must sum to one.
# keepthem equals true, ensures that the sum is still two dimensional, which is required for the division to broadcast correctly 
# in Numpy for PI. For pi There is no need to do this since it's just a 1D array.

A0 /= A0.sum(axis=1, keepdims=True)
pi0 /= pi0.sum()

A1 /= A1.sum(axis=1, keepdims=True)
pi1 /= pi1.sum()
```

```python
# log A and pi since we don't need the actual probs
logA0 = np.log(A0)
logpi0 = np.log(pi0)

logA1 = np.log(A1)
logpi1 = np.log(pi1)
```

```python
# The next step is to compute the priors will begin by counting how many samples belong to class zero and class 1
# compute priors
count0 = sum(y == 0 for y in Ytrain)
count1 = sum(y == 1 for y in Ytrain)
total = len(Ytrain)
p0 = count0 / total
p1 = count1 / total
logp0 = np.log(p0)
logp1 = np.log(p1)
p0, p1
```

```python
# we can see that P0 is about 33 percent and P1 is about sixty six percent because of this,it wouldn't make sense 
# to use the maximum likelihood method. Instead, we should look at the posterior, which corresponds to the map method.
# Note also that because the classes are slightly imbalanced, we may want to use a metric other than the accuracy to
# evaluate a classifier.
```

```python
# build a classifier
class Classifier:
    def __init__(self, logAs, logpis, logpriors):
        self.logAs = logAs
        self.logpis = logpis
        self.logpriors = logpriors
        self.K = len(logpriors) #no of classes
    
#     The next step is to define a function called compute log likelihood, this will take as input a line
#     of text represented by a list of integers and a class. the class tells us which Markov model to use 
#     since we have one for every class inside the function will begin by retrieving log and log pi, by indexing 
#     our lists by the class.
    def _compute_log_likelihood(self, input_, class_):
        logA = self.logAs[class_]
        logpi = self.logpis[class_]
        
        last_idx = None
        logprob = 0
        for idx in input_:
            if last_idx is None:
                #the first token
                logprob += logpi[idx]
                
            else:
                logprob += logA[last_idx, idx]
                
            #update last idx
            last_idx = idx
        return logprob
#    The next step is to define the predict function, the input into this function as the list of input sequences 
#     inside the function will begin by initializing an array to store the predictions.The next step is to loop through 
#     each input note that we use enumerate so that we get the index as well.Inside the loop, we compute the posterior for
#     each class using a list comprehension.This will give us a list of posteriors one for each class.The next step is to
#     obtain a prediction, which is simply the ARG max of the list of posteriors.
    def predict(self, inputs):
        predictions = np.zeros(len(inputs))
        
        for i, input_ in enumerate(inputs):
            posteriors = [self._compute_log_likelihood(input_, c) + self.logpriors[c] for c in range(self.K)]
            pred = np.argmax(posteriors)
            predictions[i] = pred
        
        return predictions
            
```

```python
# each array must be in order since classes are assumed to index these lists
clf = Classifier([logA0, logA1], [logpi0, logpi1], [logp0, logp1])
```

```python
Ptrain = clf.predict(train_text_int)
```
